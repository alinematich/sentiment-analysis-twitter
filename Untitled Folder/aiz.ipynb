```python
##Libraries

import  numpy as np
import pandas as pd
import datetime as dt
##Import Dataset
df = pd.read_excel("./data.xlsx").iloc[0:]
df = df.drop("نماد", axis=1)
df = df.drop("ارزش", axis=1)

##preprocessing
weekDays = ("Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday")
df['weekday'] = df['تاریخ'].map(lambda x: weekDays[x.weekday()])
df['تاریخ'] = df['تاریخ'].map(lambda x: dt.datetime.strptime(str(x),"%Y-%m-%d 00:00:00").strftime('%m'))
# df = df.drop("تاریخ", axis=1)

df = pd.get_dummies(df)

col = 'مقدار قیمت پایانی'

num = 17

for i in range(1,num+1):
    df[col+str(i)] = df[col].shift(i)
    
df['output'] = df[col].shift(-1)

df.dropna(inplace=True)
df.reset_index(drop=True,inplace=True)

#missing value detection
# print(df.isna().sum())

# from matplotlib import pyplot
# df[['تاریخ',col]].plot()
# pyplot.show()
df.head()
# from matplotlib import pyplot
# from pandas.tools.plotting import autocorrelation_plot
# autocorrelation_plot(df[[col]])
# pyplot.show()
```

```python
from sklearn import preprocessing;
from sklearn.model_selection import train_test_split
from  sklearn.metrics import mean_squared_error
import math

X = df[df.columns[1:len(df.columns)-1]].copy()
y = df['output'].copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

def showpred(y_test, y_pred):
    res = pd.DataFrame(data={'test':y_test.tolist(), 'pred':y_pred.tolist()})
    res['darsad'] = res.apply(lambda row: 100*(row['test'] - row['pred'])/row['test'], axis=1)
    mse=mean_squared_error(y_test,y_pred)
    print(math.sqrt(mse))
    return res.head()
```

```python
# linear regression

from sklearn import linear_model;
learner = linear_model.LinearRegression()
learner.fit(X_train,y_train)
y_pred = learner.predict(X_test)
showpred(y_test, y_pred)
```

```python
# non linear

import statsmodels.api as sm
model=sm.OLS(y_train,X_train).fit()
y_pred=model.predict(X_test)
showpred(y_test, y_pred)
```

```python
# svm regression

from sklearn.svm import SVR
clf = SVR(gamma=0.001, C=1.0, epsilon=0.2)
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
showpred(y_test, y_pred)
```

```python
# MLPregressor

from sklearn.neural_network import MLPRegressor
model = MLPRegressor(hidden_layer_sizes=10, learning_rate_init=.2, activation='relu', max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
showpred(y_test, y_pred)
```

```python
# ARIMA

from pandas import read_csv
from statsmodels.tsa.arima_model import ARIMA

series = df['output'].iloc[1000:]
X = series.values
size = int(len(X) * 0.9)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print(t,len(test), 'predicted=%f, expected=%f' % (yhat, obs))
# error = mean_squared_error(test, predictions)
# print('Test MSE: %.3f' % error)
# plot
# pyplot.plot(test)
# pyplot.plot(predictions, color='red')
# pyplot.show()
showpred(pd.Series(test), pd.Series(predictions))
```
